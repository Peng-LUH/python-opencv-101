```python
%reset -f

import cv2 as cv
from opencv_toolkit import rescale_frame

# reading video
capture = cv.VideoCapture('../../data/videos/dog.mp4')

while True:
    isTrue, frame = capture.read()
    frame_resized = rescale_frame(frame=frame, scale=0.5)
    
    cv.imshow('video', frame)
    
    if frame_resized is not None:
        cv.imshow('video resized', mat=frame_resized)
    
    if cv.waitKey(20) & 0xFF==ord('d'):
        break

capture.release()
cv.destroyAllWindows()
```
